```python
from google.colab import files
import pandas as pd
import io # Import the 'io' module

# Uploading the file
uploaded = files.upload()

# Getting the filename and content of the uploaded file
filename = list(uploaded.keys())[0]
content = uploaded[filename]

df = pd.read_csv(io.BytesIO(content)) # Creating a DataFrame named 'df'

# Displaying the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

```python
# Setting the style for the plots
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.scatterplot(x='PreventableHospitalizationsPercent', y='Race_Eth', data=df)
plt.title('Scatterplot of PreventableHospitalizationsPercent vs Race_Eth')
plt.xlabel('PreventableHospitalizationsPercent')
plt.ylabel('Race_Eth')
plt.show()
```

```python
# Boxplot:
plt.figure(figsize=(8, 4))
sns.boxplot(x='PreventableHospitalizationsPercent', y='Race_Eth', data=df)
plt.title('Boxplot of Race_Eth by PreventableHospitalizationsPercent')
plt.xlabel('PreventableHospitalizationsPercent')
plt.ylabel('Race_Eth')
plt.show()
```

```python
# Bar Chart:
category_counts = df['PatCounty'].value_counts()
plt.figure(figsize=(8, 4))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Bar chart of PatCounty')
plt.xlabel('Race_Eth')
plt.ylabel('PreventableHospitalizationsPercent')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['PreventableHospitalizationsPercent'], bins=20, kde=True)
plt.title('Histogram of PreventableHospitalizationsPercent')
plt.xlabel('PreventableHospitalizationsPercent')
plt.ylabel('Frequency')
plt.show()
```
